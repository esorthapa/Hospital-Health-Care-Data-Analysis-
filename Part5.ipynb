## Reading data from CSV

```python
import pandas as pd
file_path = (r"C:\\Users\\user\\anaconda_projects\\Data Analysis of Health Care\\Health_dataset.csv")
df = pd.read_csv(file_path)
print(df.head())
```

## Analyze records with a documented cause of death to identify potential risk factors.


```python
# filter records where the cause of death is not empty
death_cause = df.dropna(subset='Cause of Death')

count_cause = death_cause['Cause of Death'].value_counts()
print('Most common cause of Death:')
print(count_cause)
```

### Visualizing the Finding

```python
import matplotlib.pyplot as plt

plt.figure (figsize=(12, 4))
count_cause.plot(kind='bar', color='red')
plt.xticks(rotation=45)
plt.title('Top Causes of Death')
```
