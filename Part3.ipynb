```python
import pandas as pd

# Load the raw data in project
health_data = pd.read_csv("C:\\Users\\user\\anaconda_projects\\Data Analysis of Health Care\\Health_dataset.csv")
print(health_data.head())
```

```python
# Identify the most commonly diagnosed diseases among patients
disease_counts = health_data['Disease'].value_counts()
ax = disease_counts.plot(kind= 'bar', color= 'coral')
```

```python
# Get top 3 diseases
disease_count = health_data['Disease'].value_counts().nlargest(3)
print(disease_count)
```

```python
disease_count = health_data['Disease'].value_counts().nlargest(3)
total_top = disease_count.sum()
format_value = "{:,}".format(total_top)
print("Total count of the 3 most commom diseases: ", format_value)
```

```python
dynamictop = 7
disease_count = health_data['Disease'].value_counts().nlargest(dynamictop)
total_top3 = disease_count.sum()
format_value = "{:,}".format(total_top3)
description = f"Total count of the {dynamictop} most commom diseases: "
print(description, format_value)
```

```python
# Get top 3 diseases
disease_count = health_data['Disease'].value_counts().nlargest(3)
total_diseases_count = health_data['Disease'].value_counts().sum()
print(total_diseases_count)
```

```python
# Calculate top 3 percentage

disease_count3 = health_data['Disease'].value_counts().nlargest(3)
total_diseases_count = health_data['Disease'].value_counts().sum()
percentage = ( disease_count3/total_diseases_count) * 100

print(percentage)

roundpercentage = percentage.round(0).astype(int)
print(roundpercentage)
```

```python
disease_count3 = health_data['Disease'].value_counts().nlargest(3)
total_diseases_count = health_data['Disease'].value_counts().sum()
percentage = ( disease_count3/total_diseases_count) * 100

#print(percentage)

roundpercentage = percentage.round(0).astype(int)

for disease, perct in roundpercentage.items():
    print(f"{disease}: {perct}%")
```

```python
disease_counttop3 = health_data['Disease'].value_counts().nlargest(3)
total_diseasestop3count = disease_counttop3.sum()

# print(total_diseasestop3count)

pcttop3 = (disease_counttop3/total_diseasestop3count) * 100

# print(pcttop3)

roundpct = pcttop3.round(0).astype(int)
# print(roundpct)

for disease, anything in roundpct.items():
    print(f"{disease}: {anything}%")
```
